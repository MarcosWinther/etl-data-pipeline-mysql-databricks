# 🥉 Camada Bronze (Ingestão)

### ⚙ Configuração e Importação (Python)

```python
%python
# Definindo os caminhos dos arquivos CSV no DBFS
path_raiz = "/Volumes/main/default/dados_papelaria/"

tabelas = ["clientes", "produtos", "pedidos", "itens_pedido"]

print("🚀 Iniciando ingestão da camada Bronze...")
```

### 🧱 Carga dos Dados Brutos (Python)

```python
%python
for tabela in tabelas:
    try:
        # Lendo o CSV do Volume
        df = (spark.read
              .format("csv")
              .option("header", "true")
              .option("inferSchema", "true") 
              .load(f"{path_raiz}{tabela}.csv"))
        
        # Salvando na Bronze (Camada de Tabelas)
        # Se não tiver um catálogo específico, ele salva no default
        df.write.format("delta").mode("overwrite").saveAsTable(f"bronze_{tabela}")
        print(f"✅ Tabela bronze_{tabela} carregada com sucesso do Volume!")
        
    except Exception as e:
        print(f"❌ Erro ao carregar {tabela}: {e}")
```
